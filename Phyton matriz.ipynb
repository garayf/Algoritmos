```python
import matplotlib.pyplot as plt
from random import randint
import numpy as np
from time import time

grafica=[]
for j in range(40):
    A = np.zeros((10 , 10))
    for k in range(5):
        A[k]= randint(0,10)
    
    B = np.zeros((10, 10))
    
    for l in range(5):
        
        B[l]= randint(0,10)
    
    init = time()
    c = A*B
    
    end = time() - init
    
    grafica.append(end)
    
dibujo =np.array(grafica)

plt.plot(dibujo)
```
